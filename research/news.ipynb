```python
from openbb import obb
import pandas as pd
from dotenv import load_dotenv
import os

load_dotenv()

obb.account.login(pat=os.environ['PERSONAL_ACCESS_TOKEN'])
```

```python
# Get news
df = pd.DataFrame(columns=['symbols', 'title', 'url'])

news = obb.news.company(symbols="AAPL,MSFT", limit=20, provider='yfinance')

idx = 0
for article in news.results:
    values = [article.symbols, article.title, article.url]
    df.loc[idx] = values
    idx += 1
df
```
